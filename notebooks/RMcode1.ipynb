```python
users_count = dfp['user_id'].value_counts()
users_count
```

```python
#Filter out customer with less that 5 orders (still active).
active = users_count[users_count > 5]
active
```

```python
#Before that I only have the list. Now directly find active user on main data frame. 
dfp_active = dfp[dfp['user_id'].isin(active.index)]  #index is for choosing ID only
dfp_active
```

```python
import implicit
from scipy.sparse import coo_matrix
```

```python
from scipy.sparse import coo_matrix

dfp_active['user_idx'] = dfp_active['user_id'].astype('category').cat.codes
dfp_active['item_idx'] = dfp_active['product_id'].astype('category').cat.codes


df_counts = dfp_active.groupby(['item_idx', 'user_idx']).size().reset_index(name='count')
num_users = dfp_active['user_idx'].nunique()
num_items = dfp_active['item_idx'].nunique()

interaction_matrix = coo_matrix(
    (df_counts['count'], (df_counts['user_idx'], df_counts['item_idx'])),
    shape=(num_users, num_items)
).tocsr()
print("Interaction matrix shape (users x items):", interaction_matrix.shape)
```

```python
#Train model
from implicit.als import AlternatingLeastSquares

model = AlternatingLeastSquares(factors=64, regularization=0.1, iterations=20)
model.fit(interaction_matrix)
```

```python
user_idx = 0
user_items = interaction_matrix[user_idx]
recommendations = model.recommend(user_idx, interaction_matrix[user_idx], N=10)
item_ids, scores = recommendations
print("Top 10 recommended product IDs:")
print( dfp_active['product_id'].astype('category').cat.categories[ recommendations[0] ].tolist() )
```

```python
#Map back to original product_ID
item_mapping = dfp_active['product_id'].astype('category').cat.categories
recommended_product_ids = item_mapping[recommendations[0]].tolist()

print("Recommended original product_ids:", recommended_product_ids)
```
